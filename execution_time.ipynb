```python
import matplotlib.pyplot as plt
```

```python
delay_types = df['Delay Type'].unique()
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(8, 30))

# Iterate over the unique delay types and create a plot for each
for i, delay_type in enumerate(delay_types):
    # Filter the DataFrame for the current delay type
    filtered_df = df[df['Delay Type'] == delay_type]
    # Plot
    filtered_df.plot(x='Iteration', y=['Hadoop', 'Spark'], kind='bar', stacked=False, title=delay_type, ax=axes[i])

# Adjust layout
plt.tight_layout()

plt.show()
```
